## (4)基于titanic_passenger_list中的乘客信息，，1-891号乘客的国别分布情况如何?(10分)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
data = pd.read_csv('data\\titanic_passgener_list.csv', encoding="ANSI")

# 获取第1到891行的country列数据
countries = data.iloc[1:891, data.columns.get_loc('country')].values

# 统计每个国别出现的次数
unique, counts = np.unique(countries, return_counts=True)

# 输出结果
for i in range(len(unique)):
    print(unique[i], counts[i])

# 可视化结果
plt.figure(figsize=(10, 6))
plt.bar(unique, counts)
plt.title('Country Distribution')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

# (1) 请画出三种渠道的广告投放与销量的散点图。


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#解决不显示中文的问题
plt.rcParams['font.sans-serif']=['Simhei']

# 读取csv文件
data_f = r'data\Advertising.csv'
df = pd.read_csv(data_f)

# 绘制散点图
plt.scatter(df['TV'], df['Sales'], label='TV')
plt.scatter(df['Radio'], df['Sales'], label='Radio')
plt.scatter(df['Newspaper'], df['Sales'], label='Newspaper')

plt.title('三种广告投放和销量散点图')
plt.legend()
plt.xlabel('投放量')
plt.ylabel('销量')
plt.show()
```

#  （2）三种渠道的广告投放是否都可以有效促进销售? 
## 都有效，可以从散点图看出

# （3）哪种渠道对销售的促进作用最强? 

```python
# 计算皮尔逊相关系数
corr = df.corr()
print(corr)
# 生成热力图
sns.heatmap(corr, annot=True)
```
